```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
class OdeNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(OdeNN, self).__init__()
        # Initialize the layers with specified sizes
        self.fc1 = nn.Linear(input_size, hidden_size)
        #self.bn1 = nn.BatchNorm1d(hidden_size)
        #self.hidden_hidden = nn.Linear(hidden_size, hidden_size) 
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, X, Y):

        xy_combined = torch.stack((X.flatten(), Y.flatten()), dim=1)

        xy_combined = torch.sigmoid(self.fc1(xy_combined))  # Applying sigmoid activation function to the first layer
        #xy_combined = self.bn1(xy_combined)
        #xy_combined = torch.sigmoid(self.hidden_hidden(xy_combined))
        xy_combined = self.fc2(xy_combined)

        xy_combined = xy_combined.view(X.shape[0], X.shape[1],X.shape[2]) 
        return xy_combined


def calculate_boundary_loss(x_grid,y_grid,u, BC_values, gamma):
    # Left boundary
    left_boundary_loss = ((u[:, 0] - BC_values[0](x_grid[:, 0], y_grid[:, 0])) ** 2).sum()
    
    # Right boundary
    right_boundary_loss = ((u[:, -1] - BC_values[1](x_grid[:, -1], y_grid[:, -1])) ** 2).sum()

    # Bottom boundary
    bottom_boundary_loss = ((u[0, :] - BC_values[2](x_grid[0, :], y_grid[0, :])) ** 2).sum()

    # Top boundary
    top_boundary_loss = ((u[-1, :] - BC_values[3](x_grid[-1, :], y_grid[-1, :])) ** 2).sum()

    boundary_loss = gamma[0]*left_boundary_loss + gamma[1]*right_boundary_loss + gamma[2]*bottom_boundary_loss + gamma[3]*top_boundary_loss

    return boundary_loss

def poisson_PDE_loss(network_output, x_grid, y_grid, RHS, BC_values, gamma):  #add LHS_coeffs parameter,
    '''Loss function for neurally computing the solution to poisson's equation
    of the form laplacian*u=f(x,y) on a rectangular domain. Loss is computed as (laplacian(u_pred) - f(x,y))^2 + 
    gamma*(u_pred(Boundary Points)-u(Boundary Points))^2 where gamma is a weighting parameter used 
    to increase or decrease the influence of the boundary on the overall loss.

    - GRID MUST BE FORMATTED AS 'xy'!!!!! (with x point values changing from col to col)
    - Here, for Lap(u)=f(x, y), f is given by RHS, an anonymous function of x & y. i.e. for y''=2, RHS would be 
    RHS = lambda x: 2
    - BC_values is a LIST OF 4 ANONYMOUS FUNCTIONS with form [bc_xa, bc_xb, bc_yc, bc_yd]. The type of boundary conditions
    will be specified in the BC_types section.
    - xbar is an array of evaluation points (x,y) within the PDE domain, having form [[x1,y1],...,[xn,yn]]
    - For simplicity, it is assumed that all mesh points are passed through the net, not only interior pts.
    this means that the the actual 'weight' of boundary points is given by (gamma+1)
    - IGNORE: ASSUME ALL DIRICHLET BC. bc_types is an array that tells the types of the four BCs in form 
    ['type of bc_a', 'type of bc_b', 'type of bc_c', 'type of bc_d']. Options are 
    'D' for Dirichlet, or 'N' for Neumann. Types of all four BCs are set to dirichlet by default'''
    u = network_output

    du_dx = torch.autograd.grad(u, x_grid, create_graph=True,  grad_outputs=torch.ones_like(network_output), retain_graph=True, allow_unused=True)[0]
    d2u_dx2 = torch.autograd.grad(du_dx, x_grid, create_graph=True,  grad_outputs=torch.ones_like(network_output), retain_graph=True, allow_unused=True)[0]

    du_dy = torch.autograd.grad(u, y_grid, create_graph=True,  grad_outputs=torch.ones_like(network_output), retain_graph=True, allow_unused=True)[0]
    d2u_dy2 = torch.autograd.grad(du_dy, y_grid, create_graph=True,  grad_outputs=torch.ones_like(network_output), retain_graph=True, allow_unused=True)[0]

    # ODE Loss: given as a truncation error effectively
    #ode_loss = torch.mean((second_derivative - RHS(x)) ** 2)
    #ode_loss = torch.mean((d2u_dx2 + d2u_dy2 - RHS(x_grid, y_grid)) ** 2)
    ode_loss = ((d2u_dx2 + d2u_dy2 - RHS(x_grid, y_grid)) ** 2).sum()

    # Boundary Conditions Loss: (y(a) - 1)^2 and (y(b) - 1)^2
    boundary_loss = calculate_boundary_loss(x_grid, y_grid, u, BC_values, gamma)
  
    # Total Loss
    total_loss = ode_loss + boundary_loss
    return total_loss
```

```python
def train_model(model, x_grid, y_grid, criterion, optimizer, num_epochs, RHS, BCs, gamma=[1,1,1,1]):
    '''BCs is a 1x2 array of form [y(a), y(b)]
    Gamma is the additional weighting of the BC loss (gamma=0 means they are weighted the same as other points)
    RHS is the right-hand-side of the 2nd order ode of form y''=f(x), in anonymous function format'''
    #scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=10, factor=0.1, verbose=True)

    for epoch in range(num_epochs):
        # Internal points
        #x_internal = torch.cat([torch.tensor([x_bounds[0]]),torch.rand(98) * (x_bounds[1] - x_bounds[0]) + x_bounds[0],torch.tensor([x_bounds[1]])])
        #y_internal = torch.cat([torch.tensor([y_bounds[0]]),torch.rand(98) * (y_bounds[1] - y_bounds[0]) + y_bounds[0], torch.tensor([y_bounds[1]])])
        #x_g, y_g = torch.meshgrid(x_internal, y_internal, indexing='xy')
        #x_grid, y_grid = x_g.unsqueeze(-1), y_g.unsqueeze(-1)

        optimizer.zero_grad()
        x_grid.requires_grad = True
        y_grid.requires_grad = True
        u_hat = model(x_grid, y_grid)
        #boundary_output = model(boundary_points)
        loss = criterion(u_hat, x_grid, y_grid, RHS, BCs, gamma)
        loss.backward()
        optimizer.step()

        #scheduler.step(loss)
        
        if epoch % 100 == 0:
            print(f'Epoch {epoch+1}, Loss: {loss.item()}')
    return model
```

```python
#Data & Boundary Point Processing

#Region Bounds (INPUT) - to match cartesian prod [xa, xb]x[ya, yb]
x_bounds = [0,2*torch.pi]
y_bounds = [0,torch.pi]

#DATA PREPROCESSING - Gets rectangular mesh into two concatenated column vectors (one for x, one for y)
x_points = torch.linspace(x_bounds[0],x_bounds[1],100)
y_points = torch.linspace(y_bounds[0],y_bounds[1],100)
x_grid, y_grid = torch.meshgrid(x_points, y_points, indexing='xy')
[x_grid, y_grid] = [x_grid.unsqueeze(2), y_grid.unsqueeze(2)]
```

```python
#Innitializing Our ODE and Domain 
# We are looking to solve: [-lap(u)=sin(x)cos(y), on (0,1)x(5,10), where u=0 on boundary]

#ODE information
#BCs = torch.tensor([1, -1], dtype=torch.float32)
RHS = lambda x, y: -torch.sin(x)*torch.sin(y)
zero_bc_fn = lambda x, y: 0           #The way that the loss function is defined, this must be in anon form
bc_values = [zero_bc_fn, zero_bc_fn, zero_bc_fn, zero_bc_fn]

#Setting up Model
model = OdeNN(2,10,1)           #width 10 chosen here... why not?
optimizer = torch.optim.Adam(model.parameters(), 0.006)
epochs = 20000
gamma=[0.5,0.5,0.5,0.5]         #BCs weighted twice as much in this test

#Training
model = train_model(model,x_grid, y_grid, poisson_PDE_loss,optimizer=optimizer,num_epochs=epochs,RHS=RHS, BCs=bc_values, gamma=gamma)
```

```python
#Checking Performance (graphically)
#finding predicted outputs
with torch.no_grad():
    pred = model(x_grid, y_grid).squeeze(2)

#Analytical outputs (anon fn)
analyticalsoln = lambda x,y: 0.5*np.sin(x)*np.sin(y)

#Plotting
[x_plot, y_plot] = [x_grid.squeeze(2), y_grid.squeeze(2)]
fig, ax = plt.subplots(1,2,subplot_kw={"projection": "3d"},figsize=(10, 6))
#plt.plot(x_points.detach().numpy(), pred.detach().numpy(),label = 'Neural soln', marker ='*')
real_soln_pts = analyticalsoln(x_plot.detach().numpy(), y_plot.detach().numpy())
ax[0].plot_surface(x_plot.detach().numpy(), y_plot.detach().numpy(), real_soln_pts, label = 'Analytical Soln',cmap='viridis')
ax[1].plot_surface(x_plot.detach().numpy(), y_plot.detach().numpy(), pred.detach().numpy(), label = 'Analytical Soln', cmap='viridis')
ax[0].set_title('Analytical')
ax[1].set_title('Neural')
plt.xlabel('X')
plt.ylabel('Y')
fig.suptitle('Plot of Solution to $\\nabla^2 u = \sin(x)\\sin(y), u=0$ on $\\Omega$')
plt.show()
```
