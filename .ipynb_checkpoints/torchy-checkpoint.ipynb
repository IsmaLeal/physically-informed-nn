```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

## Part 1: Setting Up Neural Network & Loss Fn
### One hidden layer used for simplicity, width left variable. torch.autograd used to automatically find derivatives of tensor graph objects

```python
class OdeNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(OdeNN, self).__init__()
        # Initialize the layers with specified sizes
        self.fc1 = nn.Linear(input_size, hidden_size) 
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = torch.sigmoid(self.fc1(x))  # Applying ReLU activation function to the first layer
        x = self.fc2(x)  # Applying sigmoid activation function to the second layer
        return x

#Custom Loss function for 2nd order ODEs
def ODE_loss(network_output, x, RHS, bc_a, bc_b, gamma):
    '''Loss function for second order ODE problem, calculated as (y_pred - f(x))^2 + 
    gamma*(y_pred(Boundary)-y(Boundary))^2 where gamma is a weighting parameter used 
    to increase or decrease the influence of the boundary on the overall loss.
    
    Here, for y''=f(x), f(x) is given by RHS, an anonymous function of x. i.e. for y''=2, RHS would be 
    RHS = lambda x: 2
    
    x is a vector of evaluation points in within the ODE domain
    
    For simplicity, it is assumed that all mesh points are passed through the net, not only interior pts.
    this means that the the actual 'weight' of boundary points is given by (gamma+1) '''

    # Calculate the second derivative w.r.t inputs
    y = network_output.sum()
    first_derivative = torch.autograd.grad(y, x, create_graph=True, retain_graph=True, allow_unused=True)[0]
    second_derivative = torch.autograd.grad(first_derivative.sum(), x, create_graph=True, retain_graph=True, allow_unused=True)[0]

    # ODE Loss: given as a truncation error effectively
    ode_loss = torch.mean((second_derivative - RHS(x)) ** 2)
    
    # Boundary Conditions Loss: (y(a) - 1)^2 and (y(b) - 1)^2
    boundary_loss = gamma * ( (network_output[0] - bc_a) ** 2 + (network_output[-1] - bc_b) ** 2)
    
    # Total Loss
    total_loss = ode_loss + boundary_loss
    return total_loss
```

## Part 2: Training
### .requires_grad used to retain gradients for loss computation; This is how we tell pytorch that gradients w.r.t x are useful, and allows them be used by the aurograd method above. BCs and RHS left general.

```python
def train_model(model, x_points, criterion, optimizer, num_epochs, RHS, BCs, gamma):
    '''BCs is a 1x2 array of form [y(a), y(b)]
    Gamma is the additional weighting of the BC loss (gamma=0 means they are weighted the same as other points)
    RHS is the right-hand-side of the 2nd order ode of form y''=f(x), in anonymous function format'''
    for epoch in range(num_epochs):
        optimizer.zero_grad()
        x_points.requires_grad = True
        y_hat = model(x_points)
        loss = criterion(y_hat, x_points, RHS, BCs[0], BCs[1], gamma)
        loss.backward()
        optimizer.step()
        
        if epoch % 100 == 0:
            print(f'Epoch {epoch+1}, Loss: {loss.item()}')
    return model
```

## Part 3: Setting up Problem & Running
#### ODE given by $y''=-2, \; y(0)=y(1)=1$
#### For a 3-layer net with width 10, getting loss ~ 10e-5 

```python
#Innitializing Our ODE and Domain 
# We are looking to solve: [y''+2=0] on x in [0,1]
#We have BCs given by y(0)=y(1)=1, and we know the exact sol'n is given by y = 1 + x(1-x)

#ODE information
BCs = [1, 1]
RHS = lambda x: -2           #The way that the loss function is defined, this must be in anon form
x_points = torch.linspace(0,1,100).unsqueeze(1)     #unsqueeze used to ensure model understands that inputs are 1D

#Setting up Model
model = OdeNN(1,10,1)           #width 10 chosen here... why not?
optimizer = torch.optim.Adam(model.parameters(), 0.005)
epochs = 10000
gamma=1         #BCs weighted twice as much in this test

#Training
model = train_model(model,x_points,ODE_loss,optimizer=optimizer,num_epochs=epochs,RHS=RHS,BCs=BCs,gamma=gamma)
```

## Part 4: Plotting

```python
#Checking Performance (graphically)
#finding predicted outputs
with torch.no_grad():
    pred = model(x_points)

#Analytical outputs (anon fn)
analyticalsoln = lambda x: 1+ x*(1-x)

#Plotting
plt.plot(x_points.detach().numpy(), pred.detach().numpy(),label = 'Neural soln', marker ='*')
plt.plot(x_points.detach().numpy(),analyticalsoln(x_points.detach().numpy()),label = 'Analytical Soln, y=x*(1-x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of Analytical and Neural Solution to $y^{\prime \prime}=2$')
plt.legend()
plt.show()
```

## Part 5: Testing a Challenge ODE
#### Given by $y''=x sin(x), \; y(0)=0, \; y(2e)=\pi$

#### Analytical (*painful*) sol given by $y= -x sin(x) - 2 cos(x) + \frac{x(\pi - 2 + 2cos(2e)+2e sin(2e))}{2e}+2$

```python
#y'' = x*sin(x)
#ODE information
BCs = [0, np.pi]
RHS_challenge = lambda x: x*torch.sin(x)           #The way that the loss function is defined, this must be in anon form
x_points = torch.linspace(0,2*np.e,300).unsqueeze(1)     #unsqueeze used to ensure model understands that inputs are 1D

#Setting up Model
model_challenge = OdeNN(1,100,1)           #width 10 chosen here... why not?
optimizer_challenge = torch.optim.Adam(model_challenge.parameters(), 0.008)
epochs = 10000
gamma=1         #BCs weighted twice as much in this test

#Training
model = train_model(model_challenge,x_points,ODE_loss,optimizer=optimizer_challenge,num_epochs=epochs,RHS=RHS_challenge,BCs=BCs,gamma=gamma)
```

```python
#Checking Performance (graphically)
#finding predicted outputs
with torch.no_grad():
    pred = model_challenge(x_points)

#Analytical outputs (anon fn)
#analyticalsoln = lambda x: (1/(2*np.e))*(np.pi*x - 2*x*np.e*np.sin(x) + 2*x*np.e*np.sin(2*np.e) + 2*x*np.cos(2*np.e) - 4*np.e*np.cos(x) + 4*np.e)
analyticalsoln = lambda x: -x*np.sin(x) - 2*np.cos(x) + (x*(np.pi - 2 + 2*np.cos(2*np.e)+ 2*np.e*np.sin(2*np.e))/(2*np.e)) + 2
#Plotting
plt.plot(x_points.detach().numpy(), pred.detach().numpy(),label = 'Neural soln', marker ='*')
plt.plot(x_points.detach().numpy(),analyticalsoln(x_points.detach().numpy()),label = 'Analytical Soln')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sample Analytical & Neural Solution to $y^{\prime \prime}=x*sin(x)$')
plt.legend()
plt.show()
```
